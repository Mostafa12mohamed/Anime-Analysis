### 4] Data Analysis & Visualization

# 1) Top 10 Anime by Popularity


```python
top_10_anime = df.groupby('English').agg({'Popularity': 'sum'}).sort_values(by='Popularity', ascending=False).head(10)
top_10_anime
```

```python
#bar plot for Top 10 Anime with x-axis: English and Y-axis: Popularity
plt.figure(figsize=(10, 6))
plt.bar(top_10_anime.index, top_10_anime['Popularity'])
plt.xlabel('English')
plt.ylabel('Popularity')
plt.title('Top 10 Anime by Popularity')
plt.xticks(rotation=90)
plt.show()
```

```python
# 2)Score distribution
plt.figure(figsize=(10, 6))
plt.boxplot(df['Score'])
plt.xlabel('Score')
plt.title('Score Distribution')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Score'], bins=20, kde=True)
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.title('Score Distribution')
plt.show()
```

```python
######## 3) Yearwise Members ############
# TOP 7 years by total numbers

df['Premiered'] =  df['Premiered'].str.split(' ').str[1]
top_years = df.groupby('Premiered').agg({'Members': 'sum'}).sort_values(by='Members', ascending=False).head(7)
top_years
```

```python
plt.figure(figsize=(10, 6))
plt.bar(top_years.index, top_years['Members'])
plt.xlabel('Year')
plt.ylabel('Total Members')
plt.title('Top 7 Years by Total Members')
plt.show()
```
